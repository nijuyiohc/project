**1. 지역에 따라 선호하는 게임의 장르**

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 장르별 갯수 확인하기
games['Genre'].value_counts()
```

```python
# 게임 장르와 판매량 컬럼을 추출해 데이터프레임 만들기
genre = games.groupby(by='Genre')
sales_group={col : "sum" for col in games.columns[5:9]}
sales_by_genre = genre.agg(func=sales_group)
sales_by_genre
```

```python
# NA지역에서 선호하는 게임 장르 시각화
sales_by_genre['NA_Sales'].plot(kind = 'bar', color = 'skyblue')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Popular genre in NA')
plt.show()
```

북미 지역에서는 Action장르와 Sports, shooter장르가 선호되는것으로 보여진다.

```python
# EU지역에서 선호하는 게임 장르 시각화
sales_by_genre['EU_Sales'].plot(kind = 'bar', color = 'skyblue')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Popular genre in EU')
plt.show()
```

유럽 지역에서는 북미 지역과 똑같이 Action 장르와 Sports, shooter장르가 선호되는것으로 보여진다.

```python
# JP지역에서 선호하는 게임 장르 시각화
sales_by_genre['JP_Sales'].plot(kind = 'bar', color = 'skyblue')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Popular genre in JP')
plt.show()
```

일본에서는 위의 데이터들과 다르게 Role-playing 장르가 압도적으로 선호도가 높다고 보여진다.

```python
# Other지역에서 선호하는 게임 장르 시각화
sales_by_genre['Other_Sales'].plot(kind = 'bar', color = 'skyblue')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Popular genre in Other region')
plt.show()
```

위의 세 지역을 제외한 다른 지역들에서는 마찬가지로 Action 장르와 Sports,Shooter 장르가 선호되고 있다.

```python
plt.pcolor(sales_by_genre)
plt.xticks(np.arange(0.5, len(sales_by_genre.columns), 1), sales_by_genre.columns)
plt.yticks(np.arange(0.5, len(sales_by_genre.index), 1), sales_by_genre.index)
plt.title('Heatmap', fontsize=15)
plt.ylabel('Genre', fontsize=14)
plt.colorbar()

plt.show()
```



> **결과적으로 장르-출고량의 데이터를 바탕으로 일본 지역을 뺀 나머지 데이터상의 지역들에서는 Action 장르와 Sports, Shooter가 선호받는다는 사실을 알 수 있다.**


2. 연도별 게임의 트렌드

```python
games_df = games
```

```python
games_df["Year"]=games_df["Year"].astype('str')
games_df['Year'].dtype
```

```python
# 연도별로 출시된 게임 10년단위로 그룹화
games_1980s = games_df[games_df['Year'].str.contains('198')]
games_1990s = games_df[games_df['Year'].str.contains('199')]
games_2000s = games_df[games_df['Year'].str.contains('200')]
games_2010s = games_df[games_df['Year'].str.contains('201')]
games_1980s
```

```python
# 1980년대에 발매된 게임의 플랫폼 갯수세기
games_1980s['Platform'].value_counts()
```

```python
# value_counts값을 바탕으로 bar plot 만들기
games_1980s_by_platform = games_1980s['Platform'].value_counts().head()
sns.barplot(x=games_1980s_by_platform.index, y=games_1980s_by_platform)
plt.show()
# Value_counts값에서 나온 값의 비율로 bar plot 만들기
games_1980s_by_platform = games_1980s['Platform'].value_counts(normalize=True).head()
sns.barplot(x=games_1980s_by_platform.index, y=games_1980s_by_platform)
plt.show()
# value_counts값에서 나온 값으로 pie chart 만들기
games_1980s_by_platform = games_1980s['Platform'].value_counts().head()
plt.pie(x=games_1980s_by_platform, labels=games_1980s_by_platform.index, autopct='%.1f%%')
plt.show()
```

1980s : 5개의 플랫폼이 존재하며, 2600과 NES플랫폼이 플랫폼 대부분을 차지하고 있다.

```python
# 1990년대에 발매된 게임의 플랫폼 갯수세기
games_1990s['Platform'].value_counts()
```

```python
# value_counts값을 바탕으로 bar plot 만들기
games_1990s_by_platform = games_1990s['Platform'].value_counts().head(10)
sns.barplot(x=games_1990s_by_platform.index, y=games_1990s_by_platform)
plt.show()
# Value_counts값에서 나온 값의 비율로 bar plot 만들기
games_1990s_by_platform = games_1990s['Platform'].value_counts(normalize=True).head(10)
sns.barplot(x=games_1990s_by_platform.index, y=games_1990s_by_platform)
plt.show()
# value_counts값에서 나온 값으로 pie chart 만들기
games_1990s_by_platform = games_1990s['Platform'].value_counts().head(10)
plt.pie(x=games_1990s_by_platform, labels=games_1990s_by_platform.index, autopct='%.1f%%',)
plt.show()
```
